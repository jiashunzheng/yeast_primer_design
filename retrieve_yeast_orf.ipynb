```
def readSeq(fasta):
  seqs = {}
  pattern = re.compile('^>([^\s]*) ([^\s]*) .*')
  name1 = None
  name2 = None
  with open(fasta) as data:
    for line in data:
      match = pattern.match(line)
      if match:
        name1 = match.group(1)
        name2 = match.group(2)
        seqs.setdefault(name1, [])
        seqs.setdefault(name2, [])
      else:
        seqs[name1].append(line.strip())
        seqs[name2].append(line.strip())
  
  for name in seqs:
    seqs[name] = ''.join(seqs[name])
  return seqs
```

```
seqs = readSeq('orf_genomic_1000_all.fasta')
```

```
results = {}
for gene in met_genes:
  results[gene] = primer3.designPrimers({'SEQUENCE_ID':gene, 'SEQUENCE_TEMPLATE':seqs[gene]}, global_args)
```

```
data = []
columns = ['PRIMER_LEFT_0_SEQUENCE', 'PRIMER_LEFT_0', 'PRIMER_LEFT_0_TM', \
           'PRIMER_RIGHT_0_SEQUENCE', 'PRIMER_RIGHT_0', 'PRIMER_RIGHT_0_TM', \
           'PRIMER_PAIR_0_PRODUCT_SIZE']
for gene in results:
  res = [gene] + [results[gene][col] for col in columns]
  data.append(res)

data = pd.DataFrame(data, columns=['gene'] + columns)
```
